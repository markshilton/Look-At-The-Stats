```python
import pandas as pd
import json
import datetime
import time

#Function to parse the raw time strings into timedeltas for doing arithmetic on the times
def parse_time(raw_time):
    if raw_time == '':
        return None
    else:
        try:
            dt = datetime.datetime.strptime(raw_time, '%M:%S.%f')
            timedelta = datetime.timedelta(minutes=dt.minute, seconds=dt.second, microseconds=dt.microsecond)
            return timedelta
        except:
            dt = datetime.datetime.strptime(raw_time, '%S.%f')
            timedelta = datetime.timedelta(seconds=dt.second, microseconds=dt.microsecond)
            return timedelta

with open('seasonData.json', 'r') as jsonFile:
    seasonData = json.load(jsonFile)
    
#Just grabbing a single set of results for now. Will create a loop to do the whole lot later...    
df = pd.DataFrame.from_dict(seasonData[0]['results'][1]['categoryResults'])

#Apply the time parser to the raw times
for i in range(1,4):
    df['sector' + str(i)] = df['sector' + str(i)].map(parse_time)

#Calculate the total split times for each split
df['split1'] = df['sector1']
df['split2'] = df['sector1'] + df['sector2']
df['split3'] = df['split2'] + df['sector3']
df.head()

#Calculate the time gaps and rankings for everything

df['sector1Gap'] = df['sector1'].apply(lambda x: x - df['sector1'].min())
df['sector2Gap'] = df['sector2'].apply(lambda x: x - df['sector2'].min())
df['sector3Gap'] = df['sector3'].apply(lambda x: x - df['sector3'].min())
df['split1Gap'] = df['split1'].apply(lambda x: x - df['split1'].min())
df['split2Gap'] = df['split2'].apply(lambda x: x - df['split2'].min())
df['split3Gap'] = df['split3'].apply(lambda x: x - df['split3'].min())

df['sector1Rank'] = df['sector1'].rank()
df['sector2Rank'] = df['sector2'].rank()
df['sector3Rank'] = df['sector3'].rank()
df['split1Rank'] = df['split1'].rank()
df['split2Rank'] = df['split2'].rank()
df['split3Rank'] = df['split3'].rank()

#Now we're done with the calcs, convert the columns into more friendly formats for reading
#df['sector1'] = df['sector1'].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))

df.head()
```
